# How are grades distributed?

Let's create a USA GPA column that will be useful throughout the analysis.

```python
grades_mapper = {'A': 90, 'B': 80, 'C': 70, 'D': 60}

def calc_usa_gpa(score):
    for grade, threshold in grades_mapper.items():
        if score >= threshold:
            return grade
    return 'F'
    
data["Math GPA"] = data.apply(lambda x : calc_usa_gpa(x["math"]), axis=1)
data["Reading GPA"] = data.apply(lambda x : calc_usa_gpa(x["reading"]), axis=1)
data["Writing GPA"] = data.apply(lambda x : calc_usa_gpa(x["writing"]), axis=1)
```

Next, we are going to create a column with the number of exams the students passed

```python
def get_exams_passed(student):
    passed = 0
    for grade in ["Math GPA", "Reading GPA", "Writing GPA"]:
        if student[grade] < 'F':
            passed += 1
    return passed

data["Exams Passed"] = data.apply(lambda x : get_exams_passed(x), axis=1)
```

# Which gender did better at Math?

Before plotting sunburst charts to see which gender got better grades in which subjects, we're writing some auxiliary code

```python
def plot_sunburst_chart(subject):
    plt.figure(figsize=(30, 15))
    gpa_counts = pd.crosstab(data[f"{subject} GPA"], data["gender"])
    totals_labels = gpa_counts.index
    totals = gpa_counts.values.sum(axis=1)
    by_gender = gpa_counts.values.flatten()
    by_gender_labels = len(gpa_counts) * ["F", "M"]

    by_gender_colors = len(gpa_counts) * ['#ffb3e6', '#c2c2f0']
    explode = (0.3,0.3,0.3,0.3,0.3) 

    plt.pie(totals, 
            labels=totals_labels,
            autopct='%1.1f%%', 
            pctdistance=1.1,
            labeldistance=0.8,
            startangle=180,
            textprops={'fontsize': 22},
            radius=12)
    plt.pie(by_gender,
            labels=by_gender_labels,
            colors=by_gender_colors,
            autopct='%1.1f%%', 
            pctdistance=1.15,
            labeldistance=0.8,
            startangle=180,
            textprops={'fontsize': 24, 'color': 'white'},
            radius=7)

    centre_circle = plt.Circle((0,0), 5, color='black', fc='white',linewidth=0)
    fig = plt.gcf()
    fig.gca().add_artist(centre_circle)
    
    plt.title(f'Which gender did better at {subject}?', fontsize=24)

    plt.axis('equal')
    plt.tight_layout()
    plt.show()
plot_sunburst_chart("Math")
```

More males got As, Bs, and Cs, so they did better in math.

```python
# lets import the scipy package
import scipy.stats as stats
import math

# lets seed the random values
np.random.seed(10)

# lets take a sample size
sample_size = 1000
sample = np.random.choice(a= data['math'],
                          size = sample_size)
sample_mean = sample.mean()

# Get the z-critical value*
z_critical = stats.norm.ppf(q = 0.95)  

 # Check the z-critical value  
print("z-critical value: ",z_critical)                                

# Get the population standard deviation
pop_stdev = data['math'].std()  

# checking the margin of error
margin_of_error = z_critical * (pop_stdev/math.sqrt(sample_size)) 

# defining our confidence interval
confidence_interval = (sample_mean - margin_of_error,
                       sample_mean + margin_of_error)  

# lets print the results
print("Confidence interval:",end=" ")
print(confidence_interval)
print("True mean: {}".format(data['math'].mean()))
```

```python
# Assigning grades to the grades according to the following criteria :
# 0  - 40 marks : grade E
# 41 - 60 marks : grade D
# 60 - 70 marks : grade C
# 70 - 80 marks : grade B
# 80 - 90 marks : grade A
# 90 - 100 marks : grade O

def getgrade(percentage, status):
  if status == 'Fail':
    return 'E'
  if(percentage >= 90):
    return 'O'
  if(percentage >= 80):
    return 'A'
  if(percentage >= 70):
    return 'B'
  if(percentage >= 60):
    return 'C'
  if(percentage >= 40):
    return 'D'
  else :
    return 'E'

data['grades'] = data.apply(lambda x: getgrade(x['percentage'], x['status']), axis = 1 )

data['grades'].value_counts()
```
